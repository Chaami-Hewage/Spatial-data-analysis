```python
locations=pd.DataFrame()
```

Assign a new column to the data frame.

```python
locations["Name"]=df['location'].unique()
```

```python
locations.head()
```

Extract lat and long of data.

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator=Nominatim(user_agent='app')
```

```python
lat=[]
lon=[]
for location in locations['Name']:
  location=geolocator.geocode(location)
  if location is None:
    lat.append(np.nan)
    lon.append(np.nan)
  else:
    lat.append(location.latitude)
    lon.append(location.longitude)
```

```python
lat
```

```python
lon
```

Lets create 2 columns and add the lat and long to the dataset.

```python
locations['lat']=lat
locations['lon']=lon
```

```python
locations.head()
```
